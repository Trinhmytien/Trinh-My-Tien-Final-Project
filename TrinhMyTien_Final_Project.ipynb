**Dropping Columns ** (Dropped: 'App IQ Category', 'Stock Ticker', 'ARPDAU (Last Month, US)', 'Downloads (PoP Growth %)', 'Revenue (PoP Growth %)', 'DAU (Absolute)', 'DAU (PoP Growth)', 'DAU (PoP Growth %)', 'Platform', 'Category'); Continue to drop: 'Unified ID', 'Unified Publisher ID', 'Publisher ID', 'App ID'

```python
columns_to_drop = ['Unified ID', 'Unified Publisher ID', 'Publisher ID', 'App ID']
df = df.drop(columns=columns_to_drop)

# Display info to show the dropped columns
df.info()
```

```python
# Convert 'Date' and 'Earliest Release Date' to datetime objects
df['Date'] = pd.to_datetime(df['Date'])
df['Earliest Release Date'] = pd.to_datetime(df['Earliest Release Date'], errors='coerce')

# Convert 'RPD (All Time, WW)' to numeric
df['RPD (All Time, WW)'] = df['RPD (All Time, WW)'].replace({r'\$': ''}, regex=True).astype(float)

# Display info to show the updated data types
df.info()
```

```python
# Impute 'RPD (All Time, WW)' with the median
median_rpd = df['RPD (All Time, WW)'].median()
df['RPD (All Time, WW)'] = df['RPD (All Time, WW)'].fillna(median_rpd)

# Fill categorical columns with specified values
df['Game Sub-genre'] = df['Game Sub-genre'].fillna("Other")
df['Game Product Model'] = df['Game Product Model'].fillna("Unspecified")
df['Publisher Country'] = df['Publisher Country'].fillna("Unknown")
df['Most Popular Country by Downloads'] = df['Most Popular Country by Downloads'].fillna("Unknown")

# Dropping rows for 'Earliest Release Date' with missing values
df.dropna(subset=['Earliest Release Date'], inplace=True)

# Display info to show the updated non-null counts
df.info()
```

```python
# Select numerical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Plot histograms for numerical columns
df[numerical_cols].hist(figsize=(15, 10))
plt.tight_layout()
plt.show()
```

```python
# Select numerical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Plot box plots for numerical columns to visualize outliers
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 3, i + 1)
    sns.boxplot(y=df[col])
    plt.title(f'Box Plot of {col}')
plt.tight_layout()
plt.show()
```

```python
# Convert the splitting date to datetime
split_date = pd.to_datetime('2024-09-25')

# Create the first group: apps released before Sep 25, 2024
group1_before = df[df['Earliest Release Date'] < split_date].copy()

# Create the second group: apps released on or after Sep 25, 2024
group2_after = df[df['Earliest Release Date'] >= split_date].copy()

print("Group 1: Apps released before Sep 25, 2024")
display(group1_before.head())

print("\nGroup 2: Apps released on or after Sep 25, 2024")
display(group2_after.head())
```

```python
# Select numerical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

print("Descriptive statistics for Group 1 (Apps released before Sep 25, 2024):")
display(group1_before[numerical_cols].describe())

print("\nDescriptive statistics for Group 2 (Apps released on or after Sep 25, 2024):")
display(group2_after[numerical_cols].describe())
```

```python
# Identify object type columns
object_cols = df.select_dtypes(include='object').columns

# Standardize text formatting in object columns
for col in object_cols:
    # Trim leading/trailing spaces
    df[col] = df[col].str.strip()
    # Convert to Title Case (or .str.lower() depending on preference)
    df[col] = df[col].str.title()

print("Text formatting standardized for object columns.")
```

```python
# Display info to confirm missing values were handled
print("DataFrame Info after handling missing values:")
df.info()

print("\n" + "="*50 + "\n")

# Check unique values in a few standardized object columns
print("Unique values in 'Game Sub-genre' after standardization:")
print(df['Game Sub-genre'].unique())

print("\n" + "="*50 + "\n")

print("Unique values in 'Publisher Country' after standardization:")
print(df['Publisher Country'].unique())
```

```python
# Select numerical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Calculate the correlation matrix
correlation_matrix = df[numerical_cols].corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Variables')
plt.show()

# Select a few pairs with interesting correlation values for scatter plots
# Based on the heatmap, let's choose pairs with strong positive, strong negative, and near-zero correlations.
# Example pairs (adjust based on the actual heatmap results):
# 1. Downloads (Absolute) and Revenue (Absolute) (likely strong positive)
# 2. Downloads (Absolute) and Downloads (PoP Growth) (could be interesting, might not be strong)
# 3. Revenue (Absolute) and Revenue (PoP Growth) (could be interesting)
# 4. Downloads (Absolute) and RPD (All Time, WW) (potentially weak or no correlation)

selected_pairs = [
    ('Downloads (Absolute)', 'Revenue (Absolute)'),
    ('Downloads (Absolute)', 'Downloads (PoP Growth)'),
    ('Revenue (Absolute)', 'Revenue (PoP Growth)'),
    ('Downloads (Absolute)', 'RPD (All Time, WW)')
]

# Create scatter plots for the selected pairs
plt.figure(figsize=(15, 10))
for i, (col1, col2) in enumerate(selected_pairs):
    plt.subplot(2, 2, i + 1)
    sns.scatterplot(data=df, x=col1, y=col2)
    plt.title(f'Scatter Plot of {col1} vs {col2}')
    plt.xlabel(col1)
    plt.ylabel(col2)
plt.tight_layout()
plt.show()
```

## Bivariate analysis - categorical and numerical

### Subtask:
Generate code to explore the relationship between categorical and numerical variables using visualizations like box plots or violin plots.


**Reasoning**:
Identify key numerical and categorical columns and generate box plots to visualize the relationship between numerical and categorical variables.



```python
# Identify numerical and categorical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include='object').columns

# Exclude high cardinality categorical columns for visualization purposes
high_cardinality_cols = ['Unified Name', 'Publisher Name', 'App Name']
categorical_cols_to_plot = [col for col in categorical_cols if col not in high_cardinality_cols]

# Choose a few key numerical and categorical columns for exploration
# Based on potential interest, let's choose:
# Numerical: 'Revenue (Absolute)', 'Downloads (Absolute)', 'RPD (All Time, WW)'
# Categorical: 'Game Sub-genre', 'Game Product Model', 'Publisher Country', 'Most Popular Country by Downloads'

key_numerical_cols = ['Revenue (Absolute)', 'Downloads (Absolute)', 'RPD (All Time, WW)']
key_categorical_cols = ['Game Sub-genre', 'Game Product Model', 'Publisher Country', 'Most Popular Country by Downloads']

# Generate box plots for key numerical and categorical column pairs
for num_col in key_numerical_cols:
    for cat_col in key_categorical_cols:
        # Filter out categories with very few instances to make plots readable
        value_counts = df[cat_col].value_counts()
        valid_categories = value_counts[value_counts > 10].index # Adjust threshold as needed
        df_filtered = df[df[cat_col].isin(valid_categories)]

        if not df_filtered.empty:
            plt.figure(figsize=(15, 8))
            sns.boxplot(data=df_filtered, x=cat_col, y=num_col)
            plt.title(f'Distribution of {num_col} across {cat_col}')
            plt.xlabel(cat_col)
            plt.ylabel(num_col)
            plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for readability
            plt.tight_layout()
            plt.show()
        else:
            print(f"No data to plot for {num_col} vs {cat_col} after filtering.")
```

## Multivariate analysis

### Subtask:
Generate code for advanced visualizations to explore patterns and relationships across multiple variables. This could involve pair plots or faceted plots based on categorical variables.


**Reasoning**:
Select relevant numerical and categorical columns for multivariate analysis and generate a pair plot to visualize relationships between numerical variables colored by a categorical variable. Given the dataset size, sampling is necessary for pairplot.



```python
# Select a subset of numerical and categorical columns for multivariate analysis
numerical_cols_for_multi = ['Revenue (Absolute)', 'Downloads (Absolute)', 'RPD (All Time, WW)']
categorical_col_for_hue = 'Game Product Model' # Choose a categorical column with manageable unique values

# Sample the dataset for pairplot if it's too large
# Let's sample 1000 rows for the pairplot to keep it manageable
df_sampled = df.sample(n=1000, random_state=42).copy()

# Generate pairplot on selected numerical columns, using a categorical column for hue
sns.pairplot(df_sampled, vars=numerical_cols_for_multi, hue=categorical_col_for_hue, diag_kind='kde')
plt.suptitle('Pair Plot of Key Numerical Variables by Game Product Model (Sampled)', y=1.02)
plt.show()
```

**Reasoning**:
Generate faceted plots to explore the relationship between two numerical variables conditioned on the values of categorical variables.



```python
# Generate faceted plots for relationships between numerical variables conditioned on categorical variables
# Example: Downloads (Absolute) vs Revenue (Absolute) faceted by 'Game Product Model'

# Choose numerical and categorical columns for faceted plots
num_col_x = 'Downloads (Absolute)'
num_col_y = 'Revenue (Absolute)'
cat_col_facet = 'Game Product Model'

# Use FacetGrid for more control
g = sns.FacetGrid(df, col=cat_col_facet, col_wrap=3, height=4, aspect=1.2, sharey=False)
g.map(sns.scatterplot, num_col_x, num_col_y, alpha=0.7)
g.add_legend()
g.fig.suptitle(f'Scatter Plot of {num_col_y} vs {num_col_x} by {cat_col_facet}', y=1.02)
plt.tight_layout()
plt.show()

# Example: RPD (All Time, WW) vs Revenue (Absolute) faceted by 'Publisher Country' (top N countries)
# Choose a categorical column with more unique values and select top N for clarity
cat_col_facet_large = 'Publisher Country'
top_n_countries = df[cat_col_facet_large].value_counts().nlargest(6).index.tolist() # Select top 6 countries

df_filtered_countries = df[df[cat_col_facet_large].isin(top_n_countries)].copy()

g = sns.FacetGrid(df_filtered_countries, col=cat_col_facet_large, col_wrap=3, height=4, aspect=1.2, sharey=False)
g.map(sns.scatterplot, 'RPD (All Time, WW)', 'Revenue (Absolute)', alpha=0.7)
g.add_legend()
g.fig.suptitle(f'Scatter Plot of Revenue (Absolute) vs RPD (All Time, WW) by Top {len(top_n_countries)} Publisher Countries', y=1.02)
plt.tight_layout()
plt.show()
```
